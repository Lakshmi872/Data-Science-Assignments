```python
# now to treat the outliers we have :- 
# capping :- Replacing outlier values 
# in capping all outlier values will be replaced by upper extreme or lower extreme values
```

```python
df['LB'].quantile(0.25)   # lower quartile:-q1   it means that there are 25 % values which are less than 126
```

```python
df['LB'].quantile(0.75)     # upper quartile:-q3   it means that there are 75 % values which are less thn 140 , only 25 % values are greater thn 140
```

```python
# outlier Detection :- user defined function to calculate upper extreme and lower extreme values
```

```python
def outlier_detection(data,colname):
    q1 = data[colname].quantile(0.25)
    q3 = data[colname].quantile(0.75)
    iqr = q3-q1

    upper_extreme = q3 + (1.5*iqr)
    lower_extreme = q1 - (1.5*iqr)

    return print('lower_extreme:-',lower_extreme,', upper_extreme:-',upper_extreme,', q1:-',q1,', q3:-',q3)
```
